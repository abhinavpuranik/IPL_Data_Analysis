```python
#IPL Dataset analysis using Python Numpy Library

import numpy as np
```

```python
deliveries = np.genfromtxt('deliveries.csv', delimiter=',',names = True, dtype = None, encoding='utf-8', filling_values=0)


matches = np.genfromtxt('matches.csv', delimiter=',',names = True, encoding='utf-8', filling_values=0, invalid_raise=False)
print(matches)
```

```python
match_id = deliveries['match_id']
print(match_id)
batting_team = deliveries['batting_team']
print(batting_team)
batter = deliveries['batter']
print(batter)
bowler = deliveries['bowler']
print(bowler)
batsman_runs = deliveries['batsman_runs']
print(batsman_runs)
over = deliveries['over']
print(over)
```

```python
unique_batters = np.unique(batter)
```

```python
total_runs = []

for b in unique_batters:
    mask = batter == b
    runs = batsman_runs[mask].sum()
    total_runs.append(runs)
    print(runs)

total_runs = np.array(total_runs)
print(total_runs)


```

```python
sorted_indices = np.argsort(total_runs)[::-1]
top_5_indices = sorted_indices[:5]
top_5 = [(unique_batters[i], total_runs[i]) for i in top_5_indices]
print(top_5)
```

```python
strike_rates = []
for b in unique_batters:
    mask = batter == b
    number_balls = np.sum(mask)
    #print(number_balls)
    index = np.where(unique_batters == b)[0][0]
    strike_rate = (total_runs[index] / number_balls) * 100
    strike_rates.append(strike_rate)

strike_rates = np.array(strike_rates)
print(strike_rates)


    
```

```python
#task 3 without loops
unique_batters, inverse = np.unique(batter, return_inverse=True)

total_runs = np.zeros(len(unique_batters))
np.add.at(total_runs, inverse, batsman_runs)

balls = np.zeros(len(unique_batters))
np.add.at(balls, inverse, 1)

strike_rates = (total_runs / balls) * 100
print(strike_rates)
```

```python
#task4

# unique_bowlers, inverse = np.unique(bowler, return_inverse=True)
# total_balls = np.zeros(len(unique_bowlers))
# np.add.at(total_balls, inverse, 1)
# total_overs = total_balls / 6

# total_runs_bowler = np.zeros(len(unique_bowlers))
# np.add.at(total_runs_bowler, inverse, batsman_runs)

# economy = total_runs_bowler / total_overs
# print(economy)

total_runs = deliveries['total_runs'].astype(int)
extras_type = deliveries['extras_type']

unique_bowlers, inverse = np.unique(bowler, return_inverse = True)

runs_conceded = np.zeros(len(unique_bowlers))
np.add.at(runs_conceded, inverse, total_runs)

valid_ball_mask = (extras_type != 'wides')

balls_bowler = np.zeros(len(unique_bowlers))
np.add.at(balls_bowler, inverse[valid_ball_mask], 1)

valid = balls_bowler > 0

economy = np.zeros(len(unique_bowlers), dtype=float)
economy[valid] = runs_conceded[valid] / (balls_bowler[valid] / 6)
print(economy)

```

```python
#task5
unique_overs, inverse = np.unique(over, return_inverse= True)
runs_over = np.zeros(len(unique_overs))


np.add.at(runs_over, inverse, batsman_runs)
counts = np.zeros(len(unique_overs))
np.add.at(counts, inverse, 1)

average_runs_over = runs_over/counts
print(average_runs_over)


```

```python
#task6
four_mask = batsman_runs == 4
six_mask = batsman_runs == 6
boundary_mask = (batsman_runs == 4) | (batsman_runs == 6)
batting_team = deliveries['batting_team']
unique_batting_team, inverse = np.unique(batting_team, return_inverse=True)
boundaries_scored = np.zeros(len(unique_batting_team))
np.add.at(boundaries_scored, inverse[boundary_mask], 1)
#print(boundaries_scored)

sorted_boundaries_indices = np.argsort(boundaries_scored)[::-1]
top_5_boundary_indices = sorted_boundaries_indices[:5]
top_5_teams_boundaries = [(unique_batting_team[i], boundaries_scored[i]) for i in top_5_boundary_indices]
print(top_5_teams_boundaries)



```

```python
#task8
match_id = deliveries['match_id']
unique_matches, inverse =  np.unique(match_id, return_inverse= True)
matches_runs = np.zeros(len(unique_matches))
np.add.at(matches_runs, inverse, total_runs)
top_match_runs = np.argsort(matches_runs)[::-1]
top_match_runs_index = top_match_runs[:1]
top_runs_match = [(unique_matches[top_match_runs_index], matches_runs[top_match_runs_index])]
print(top_runs_match)
```

```python
#task9
keys = list(zip(match_id, batting_team))
unique_keys, inverse = np.unique(keys, axis = 0, return_inverse=True)

runs_per_team_match = np.zeros(len(unique_keys))
np.add.at(runs_per_team_match, inverse, batsman_runs)

match_ids_unique = unique_keys[:, 0]
teams_unique = unique_keys[:, 1]
```

```python
from collections import defaultdict
match_results = defaultdict(list)

for i in range(len(unique_keys)) :
    match = match_ids_unique[i]
    team = teams_unique[i]
    runs = runs_per_team_match[i]

    match_results[match].append((team, runs))

winners = {}

for match, team_data in match_results.items() :
    if len(team_data) == 2:
        team1, runs1 = team_data[0]
        team2, runs2 = team_data[1]

        if runs1 > runs2:
            winners[match] = team1
        elif runs2 > runs1:
            winners[match] = team2
        else:
            winners[match] = 'Tie'

print(winners)
```

```python
#task10
match_ids_matches = matches['id']
toss_winners = matches['toss_winner']

sort_idx = np.argsort(match_ids_matches)
sorted_match_ids = match_ids_matches[sort_idx]
sorted_toss_winners = toss_winners[sort_idx]
#sorting has been done for fast lookup

match_ids_deliveries = deliveries['match_id']

indices = np.searchsorted(sorted_match_ids, match_ids_deliveries)
# Step 1: Clip indices to valid range
indices = np.clip(indices, 0, len(sorted_match_ids) - 1)

# Step 2: Verify actual match (CRUCIAL)
valid_mask = sorted_match_ids[indices] == match_ids_deliveries

toss_winner_per_delivery = np.empty_like(match_ids_deliveries, dtype=sorted_toss_winners.dtype)

toss_winner_per_delivery[valid_mask] = sorted_toss_winners[indices[valid_mask]]
keys = list(zip(match_ids_deliveries, deliveries['batting_team']))

unique_keys, inverse = np.unique(keys, axis = 0, return_inverse=True)

runs = np.zeros(len(unique_keys))

np.add.at(runs, inverse, batsman_runs)

match_ids_unique = unique_keys[:, 0]
teams_unique = unique_keys[:, 1]


```

```python
from collections import defaultdict

match_data = defaultdict(list)

# Group runs per matchq
for i in range(len(unique_keys)):
    match = match_ids_unique[i]
    team = teams_unique[i]
    run = runs[i]
    
    match_data[match].append((team, run))

print(match_data)
toss_winner_more_runs = 0
total_matches = 0

for match, teams in match_data.items():
    if len(teams) != 2:
        continue
    
    (team1, run1), (team2, run2) = teams
    
    # get toss winner for this match
    toss_map = dict(zip(sorted_match_ids, sorted_toss_winners))

    toss_winner = toss_map.get(match, None)
    # print(team1, team2)
    # print(toss_winner)
    # compare
    if team1 == toss_winner:
        print("in if condition")
        #print(team1, toss_winner)
        toss_runs = run1
        opp_runs = run2
    else:
        print("in else condition")
        print(team1, toss_winner)
        toss_runs = run2
        opp_runs = run1
    print(toss_runs, opp_runs)
    if toss_runs > opp_runs:
        toss_winner_more_runs += 1

    total_matches += 1


print("Matches analyzed:", total_matches)
print("Toss winner scored more in:", toss_winner_more_runs)
print("Percentage:", (toss_winner_more_runs / total_matches) * 100)
```
